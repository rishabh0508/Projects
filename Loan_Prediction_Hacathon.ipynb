```python
train_df.drop(["Loan_Status"], axis=1, inplace=True)
train_df
```

```python
concat_df = pd.concat([train_df, test_df])
concat_df
```

```python
concat_df.info()
```

```python
import numpy as np
for i in set(["Gender", "Married", "Dependents", "Self_Employed", "Loan_Amount_Term", "Credit_History"]):
  concat_df[i].replace(np.nan,concat_df[i].mode()[0], inplace=True)
concat_df["LoanAmount"].replace(np.nan,concat_df["LoanAmount"].mean(), inplace=True)
concat_df.info()
```

```python
for i in concat_df.columns:
  if i not in set(["Loan_ID", "ApplicantIncome", "CoapplicantIncome", "LoanAmount"]):
    concat_df[i] = le.fit_transform(concat_df[i])
concat_df
```

```python
loan_id = concat_df["Loan_ID"].iloc[614:]
concat_df.drop(["Loan_ID"],axis=1, inplace=True)
x_train = concat_df.iloc[:614, :]
x_test = concat_df.iloc[614:, :]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
model = RandomForestClassifier()
model.fit(x_train, y_train)
print(accuracy_score(model.predict(x_train), y_train))
pred = model.predict(x_test)
print(pred)
pred=le1.inverse_transform(pred)
print(pred)
df = pd.DataFrame(pred)
df.to_csv("submit.csv")
```
